### Importing and Preprocessing

```python
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv') 
```

```python
# shapes of data
print('Movies -',movies.shape)
print('Credits -',credits.shape)
```

```python
# merge datasets
movies = movies.merge(credits, left_on='id',right_on='movie_id')
```

```python
movies.head()
```

```python
#shape of data
movies = movies[['movie_id','title_x','overview','genres','keywords','cast','crew']]
```

```python
#rename title_x
movies = movies.rename(columns={'title_x':'title'})
```

```python
movies.head()
```

```python
# Check for missing values
movies.isna().sum()
```

```python
# drop the na values
movies.dropna(inplace=True)
```

### feature enginering

```python
# function to convert json data into a list of dictionaries
import ast

def convert(text):
    return [i['name'] for i in ast.literal_eval(text)]
```

```python
# apply covert function to genres,keywords and cast
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
# keep only first 5 names of cast
movies['cast'] = movies['cast'].apply(lambda x:x[0:5])
movies.cast[0]
```

```python
# function to fetch director's name
def fetch_director(text):
    return [i['name'] for i in ast.literal_eval(text) if i['job'] == 'Director']

#apply function
movies['crew'] = movies['crew'].apply(fetch_director)
movies.crew[0]
```

```python
# Removing empty spaces from names to make it unique 
def collapse(lst):
    return [s.replace(" ", "") for s in lst]

for col in ['cast', 'crew', 'genres', 'keywords']:
    movies[col] = movies[col].apply(collapse)
```

```python
movies.head(3)
```

```python
# combine all into new columns with tags and remove the rest
movies['overview'] = movies['overview'].apply(lambda x: str(x).split())
movies['tags'] = (movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']).apply(lambda x: " ".join(x))
movies.drop(columns=['overview','genres','keywords','cast','crew'],inplace=True)
movies.head()
```
